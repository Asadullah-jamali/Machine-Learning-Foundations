```python
#target
t=[0,1,1,1]

#values of w1 and w2 (fluid power or weights)
w=[0.1,0.2]

lr=0.1

# values of x1 and x2 on different pattrens
x=[[0,0],[0,1],[1,0],[1,1]]

#threshold value
th=0.5

#count the no of ok reuslts in each check section
okcount=0

# check section repeation
for k in range(0,5):

    # no of checks
    print('check no',k+1)

    # repeation of pattrens in each check
    for i in range(0,4):

        # pre assigned value of net before calculation (zero for each pattren check)
        net = 0

        # calculating net value by puting x1 and x2 and w1 and w2 values
        for j in range(0,2):

            # for net = w1*x1 + w2*x2   (value of net changes because of w1 and w2 values changes in end if error found)
            net+=w[j]*x[i][j]


            # for removing binary error in sum (removing binary error means if we calculate 0.2 + 0.1 the answer is not 0.3 but 0.30000000000000004)
            net= round(net,1)

        # checking for error value using step function
        if net >= th:
            y=1
        else:
            y=0

        # error value
        error=t[i]-y

        # if error found changing the values of w1 and w2 using formula w1 = w1 + x1*error*lr and w2 = w2 + x2*error*lr
        for j in range(0,2):

            # if error found changing the value of w1 and w2
            if error == 1:
                a=round((x[i][j]*error*lr),1)
                w[j]=round(a+w[j],1)

            # does not change values of w1 and w2 if error does not found


        # check status 
        if error == 0:
            status = 'ok'

            # for counting no of ok in each check section if it reach 4 it means all result are ok 
            okcount+=1
        else:
            status = 'not ok'


        # printing each pattren and check results
        print(f'pattren ={i+1} \t net = {net} \t error = {error} \t x1 = {x[i][0]} \t x2 = {x[i][1]} \t w{1} = {w[0]} \t w{2} = {w[1]} \t status = {status}')


    # if count reaches 4 it means all pattrens are clear and learning completed
    if okcount == 4:
        print('\nall done')

        
    else:

        # if any pattren result is not ok in any check then recounting from zero
        okcount=0
    print('\n\n')

        
```
